```python
from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession
config = ConfigProto()
config.gpu_options.allow_growth = True
session = InteractiveSession(config=config)
InteractiveSession.close(session)
```

```python
from tensorflow.python.framework import graph_util
from tensorflow.python.platform import gfile
```

```python
h5_path='/content/drive/Shared drives/shared/train_gray.h5'
h5_file=h5py.File(h5_path,'r')
X_train=np.array(h5_file['X_train']
# X_test=np.array(h5_file['X_test'][:1000])
```

```python
meta_path = '/content/drive/My Drive/result of trains/my_model_gray_5000_sparse_4.ckpt.meta'
ckpt_path = '/content/drive/My Drive/result of trains/my_model_gray_5000_sparse_4.ckpt'
save_path = '/content/drive/My Drive/dataset'
save_name = os.path.join(save_path,'visualize_model.pb')

with tf.Session() as sess:

    saver = tf.train.import_meta_graph(meta_path, clear_devices=True)
    # import graph把模型的计算图加载出来   
    saver.restore(sess,ckpt_path)  
    
    graph=tf.get_default_graph()
    #     导出当前计算图的GraphDef部分

    graph_def = tf.get_default_graph().as_graph_def()

    for node in graph_def.node:
        print (node.name)
        
    #     保存指定的节点，并将节点值保存为常数
    save_graph = graph_util.convert_variables_to_constants(sess, graph_def,
                                                           ["Placeholder","conv_0/BiasAdd","act_0","batch_norm_0/FusedBatchNormV3","conv_1/BiasAdd","act_1","batch_norm_1/FusedBatchNormV3","encoder_output/Sigmoid","outputs_op/Sigmoid"])
    with tf.gfile.FastGFile(save_name, mode='wb') as f:
        f.write(save_graph.SerializeToString())
        print("pd file saved successfully to: ",save_name)
```
